```python
# Movie Recommendation System using TMDB dataset + mood

import pandas as pd
import ast
from google.colab import files
uploaded = files.upload()


# Load the dataset
movies = pd.read_csv("tmdb_5000_movies.csv")

# Function to extract names from JSON-like columns
def get_names(text):
    try:
        items = ast.literal_eval(text)
        return [item["name"].lower() for item in items]
    except:
        return []

# turn genres and keywords into readable lists
movies["genre_list"] = movies["genres"].apply(get_names)
movies["keyword_list"] = movies["keywords"].apply(get_names)

# Create moods based on genre and keywords
def assign_mood(row):
    genres = row["genre_list"]
    keywords = row["keyword_list"]

    if "comedy" in genres or "family" in genres or "animation" in genres:
        return "happy"
    elif "romance" in genres:
        return "romantic"
    elif "horror" in genres or "thriller" in genres:
        return "scary"
    elif "action" in genres or "adventure" in genres:
        return "exciting"
    elif "drama" in genres or "history" in genres:
        return "emotional"
    elif "science fiction" in genres or "fantasy" in genres:
        return "imaginative"
    elif "mystery" in genres or "crime" in genres:
        return "suspenseful"
    elif "music" in genres:
        return "musical"
    else:
        return "general"

movies["mood"] = movies.apply(assign_mood, axis=1)

# Recommend movies based on genre and mood
def recommend_movies(genre, mood):
    genre = genre.lower()
    mood = mood.lower()

    recommendations = movies[
        (movies["genre_list"].apply(lambda x: genre in x)) &
        (movies["mood"] == mood)
    ]

    # If no exact match, use only the genre
    if recommendations.empty:
        recommendations = movies[
            movies["genre_list"].apply(lambda x: genre in x)
        ]

    # Sort by rating and popularity
    recommendations = recommendations.sort_values(
        by=["vote_average", "popularity"],
        ascending=False
    )

    if recommendations.empty:
        print("Sorry, no movies found.")
    else:
        print("\nRecommended movies:\n")
        for _, row in recommendations.head(10).iterrows():
            print(
                row["title"],
                "| Mood:", row["mood"],
                "| Rating:", row["vote_average"],
                "| Genres:", ", ".join(row["genre_list"])
            )

# options
print("Example genres: action, comedy, romance, drama, horror, science fiction, fantasy")
print("Example moods: happy, romantic, scary, exciting, emotional, imaginative, suspenseful")

# User text input
user_genre = input("\nWhat genre do you like? ")
user_mood = input("What mood are you in? ")

recommend_movies(user_genre, user_mood)
```
