```python
import random

nouns = ["heart", "sky", "dream", "rain", "sun"]
verbs = ["shines", "falls", "dances", "whispers", "glows"]
adjectives = ["soft", "bright", "quiet", "warm", "lonely"]

def generate_poem(description):
    line1 = "The " + random.choice(adjectives) + " " + random.choice(nouns) + " " + random.choice(verbs)
    line2 = "It feels like " + description
    line3 = "In a " + random.choice(adjectives) + " " + random.choice(nouns)
    line4 = "Everything slowly " + random.choice(verbs)

    poem = line1 + "\n" + line2 + "\n" + line3 + "\n" + line4
    return poem
```

```python
description = input("Describe your feeling or scene: ")# generate poem

poem = generate_poem(description)

print("\nYour poem:\n")
print(poem)
```
