```python
import random

# Dataset with some poem words
themes = {
    "love": {
        "nouns": ["heart", "dream", "rose", "moonlight", "song"],
        "adjectives": ["gentle", "warm", "endless", "soft", "bright"],
        "verbs": ["whispers", "blooms", "shines", "waits", "dances"]
    },
    "nature": {
        "nouns": ["river", "forest", "sky", "flower", "mountain"],
        "adjectives": ["green", "silent", "golden", "peaceful", "wild"],
        "verbs": ["flows", "sings", "grows", "glows", "breathes"]
    },
    "sadness": {
        "nouns": ["rain", "shadow", "memory", "night", "tear"],
        "adjectives": ["lonely", "gray", "quiet", "cold", "fading"],
        "verbs": ["falls", "waits", "echoes", "drifts", "sleeps"]
    },
    "hope": {
        "nouns": ["sunrise", "path", "star", "light", "future"],
        "adjectives": ["new", "brave", "golden", "open", "bright"],
        "verbs": ["rises", "guides", "shines", "calls", "begins"]
    }
}
```

```python
def detect_theme(description):
    description = description.lower()

    for theme in themes:
        if theme in description:
            return theme

    if "happy" in description or "dream" in description:
        return "hope"
    elif "tree" in description or "ocean" in description or "garden" in description:
        return "nature"
    elif "alone" in description or "cry" in description:
        return "sadness"
    else:
        return random.choice(list(themes.keys()))
```

```python
def generate_poem(description):
    theme = detect_theme(description)
    words = themes[theme]

    poem = []
    poem.append(f"A {random.choice(words['adjectives'])} {random.choice(words['nouns'])} {random.choice(words['verbs'])} tonight,")
    poem.append(f"Under a {random.choice(words['adjectives'])} sky of light,")
    poem.append(f"The {random.choice(words['nouns'])} {random.choice(words['verbs'])} through the air,")
    poem.append(f"Leaving {random.choice(words['adjectives'])} dreams everywhere.")

    return theme, "\n".join(poem)

# the poem generator text box
description = input("Enter a short description for your poem: ")

theme, poem = generate_poem(description)

print("\nDetected Theme:", theme)
print("\nGenerated Poem:\n")
print(poem)
```
