```python
import pandas as pd

# small dataset
data = {
    "title": ["Toy Story", "Titanic", "Avengers", "The Notebook", "Jurassic Park"],
    "genre": ["animation", "romance", "action", "romance", "action"]
}

movies = pd.DataFrame(data)

def recommend_movie(movie_name):
    movie_name = movie_name.lower()

    # find the movie
    selected = movies[movies["title"].str.lower() == movie_name]

    if selected.empty:
        print("Movie not found.")
        return

    genre = selected.iloc[0]["genre"]

    # recommend movies with same genre
    recommendations = movies[movies["genre"] == genre]
    recommendations = recommendations[recommendations["title"].str.lower() != movie_name]

    print("\nRecommended movies:")
    for movie in recommendations["title"]:
        print("-", movie)

# user input
movie_name = input("Enter a movie you like: ")

recommend_movie(movie_name)
```
