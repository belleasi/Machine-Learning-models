```python
import pandas as pd


# Small movie dataset
data = {
    "title": [
        "Toy Story", "Finding Nemo", "Titanic", "The Notebook",
        "Avengers", "Black Panther", "Jurassic Park", "Interstellar",
        "La La Land", "Coco"
    ],
    "genre": [
        "animation", "animation", "romance", "romance",
        "action", "action", "adventure", "sci-fi",
        "musical", "animation"
    ],
"mood": [
        "happy", "happy", "sad", "romantic",
        "excited", "powerful", "thrilling", "thoughtful",
        "romantic", "emotional"
    ],
    "rating": [
        8.3, 8.2, 7.9, 7.8,
        8.0, 7.3, 8.2, 8.7,
        8.0, 8.4
    ]
}

movies = pd.DataFrame(data)

def recommend_movies(genre, mood):
    genre = genre.lower()
    mood = mood.lower()

    # Filter movies that match the user's genre and mood
    recommendations = movies[
        (movies["genre"] == genre) &
        (movies["mood"] == mood)
    ]

    # If no  match, recommend by genre only
    if recommendations.empty:
        recommendations = movies[movies["genre"] == genre]

    # rating from highest to lowest
    recommendations = recommendations.sort_values(by="rating", ascending=False)

    if recommendations.empty:
        print("Sorry, no recommendations found.")
    else:
        print("\nRecommended movies:\n")
        for index, row in recommendations.iterrows():
            print(row["title"], "-", row["genre"], "-", row["mood"], "- Rating:", row["rating"])

# User input text
print("Available genres:", movies["genre"].unique())
print("Available moods:", movies["mood"].unique())

user_genre = input("\nWhat genre do you like? ")
user_mood = input("What mood do you want? ")

recommend_movies(user_genre, user_mood)
```
